Corpus used - [`"The 18th Letter"`](https://genius.com/Rakim-the-18th-letter-always-and-forever-lyrics) (Hiphop song by Rakim)

Vocabulary size = 265 (small to medium sized corpus)

Similarity of cells, compared to collab file shared by ma'am are highlighted in top comments of each cell.

```python
#THIS CELL IS IMPORTING LIBRARIES (Same as collab shared by ma'am)

# import tensorflow as tf                  #<------- USING TENSORFLOW 2.0 IS CAUSING AN ERROR. HENCE USING TF VERSION 1.
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

```

```python
#IMPORTING DATA: import data from text file corpus (New edits in this section)
from google.colab import files
uploaded = files.upload()
parsedUpload = uploaded["corpus.txt"].decode("utf-8")

#DATA PRE-PROCESSING: Converting to lower-case

corpus = parsedUpload.lower()

parsedWords = []
for parsedWord in corpus.split():
  if parsedWord != ',' and parsedWord != '?':   # lyrics don't have '.' (period) rather they have ','(commas) and '?' (question marks) 
    parsedWords.append(parsedWord)

words = []
words = set(parsedWords)

print(parsedWords)
```

```python
#CREATING SENTENCES AS A COMBINATION OF WORDS. (Same as by ma'am)

word2int = {}                                       #creating a Dictionary
int2word = {}
vocab_size = len(words)                             # Gives the total number of unique words

for i,word in enumerate(words):                     #enumerate() method for adding a 'counter' to "words"; starting from 0               
    word2int[word] = i
    int2word[i] = word

print(word2int)
print(int2word)
```

```python
# New edits in this section 

raw_sentences = corpus.split(',')               # Raw sentences is a list of sentences #split() method splits a string into "list" on basis of separator ','
print(raw_sentences)
sentences = []
for sentence in raw_sentences:
    sentences.append(sentence.replace('?','').split())  # Remove '?' from raw sentence by replacing it with '' (empty character). Then spliting based on spaces
print(sentences)
```

```python
#MAKING POSITIVE/CONTEXT WORD-PAIRS WITHIN A GIVEN WINDOW SIZE. (Same)

WINDOW_SIZE = 2

data = []
for sentence in sentences:
    for word_index, word in enumerate(sentence):
        for nb_word in sentence[max(word_index - WINDOW_SIZE, 0) : min(word_index + WINDOW_SIZE, len(sentence)) + 1] :    #making neighbor-words pair.
            if nb_word != word:
                data.append([word, nb_word])
print(data)
```

```python
# FUNCTION TO CONVERT WORDS TO 1-HOT VECTORS (Same)

import numpy as np

def to_one_hot(data_point_index, vocab_size):
    temp = np.zeros(vocab_size)                 #np.zeros() will return an Array filled with zeros.
    temp[data_point_index] = 1
    return temp

x_train = []                                    # input word
y_train = []                                    # output word

for data_word in data:
    print(data_word)
    x_train.append(to_one_hot(word2int[ data_word[0] ], vocab_size))
    y_train.append(to_one_hot(word2int[ data_word[1] ], vocab_size))
    
# convert them to numpy arrays
x_train = np.asarray(x_train)
y_train = np.asarray(y_train)


print(x_train[0])
print(y_train[0])
print(x_train.shape)
print(y_train.shape)
```

```python
# MAKING x_train and y_train (Same)

x_input = tf.placeholder(tf.float32, shape=(None, vocab_size))
y_label = tf.placeholder(tf.float32, shape=(None, vocab_size))
```

```python
# CREATING OUR NEURAL NETWORK: W1, b1, W2, b2 (Same)

EMBEDDING_DIM = 5                                                      

W1 = tf.Variable(tf.random_normal([vocab_size, EMBEDDING_DIM]))
b1 = tf.Variable(tf.random_normal([EMBEDDING_DIM]))                      #bias
hidden_representation = tf.add(tf.matmul(x_input,W1), b1)               

W2 = tf.Variable(tf.random_normal([EMBEDDING_DIM, vocab_size]))
b2 = tf.Variable(tf.random_normal([vocab_size]))
prediction = tf.nn.softmax(tf.add( tf.matmul(hidden_representation, W2), b2))
```

```python
# CREATING OUR NEURAL NETWORK: LOSS FUNCTION, LEARNING ALGO (GD)

sess = tf.Session()
init = tf.global_variables_initializer()         # Hold the values of variables 'as I told you' (e.g. Random Normal)
sess.run(init)                                                         

cross_entropy_loss = tf.reduce_mean(-tf.reduce_sum(y_train * tf.log(prediction), reduction_indices=[1]))  # The loss function

# define the training step:
train_step = tf.train.GradientDescentOptimizer(0.1).minimize(cross_entropy_loss)
```

```python
# TRAINING (Increased training iterations)

n = 10000             # train for n iterations

for _ in range(n):
    sess.run(train_step, feed_dict={x_input: x_train, y_label: y_train})
    print('loss is : ', sess.run(cross_entropy_loss, feed_dict={x_input: x_train, y_label: y_train}))
```

```python
# Each row is a real valued vector of 5 terms, representing the words in the vocabulary on the basis of similarity in meaning (Word embeddings)
# Number of Columns = 5 (Size of Hidden Layer)
# Number of Rows = 265 (Vocabulary Size)

vectors = sess.run(W1 + b1)
print(vocab_size)     # Vocabulary Size, also the number of columns         
print(vectors)
```
